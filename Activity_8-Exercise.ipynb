```python
# 1. Library Book Management System

class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.available = True

    def borrow(self):
        if self.available:
            self.available = False
            return f'"{self.title}" has been successfully borrowed.'
        return f'"{self.title}" is already borrowed.'

    def return_book(self):
        if not self.available:
            self.available = True
            return f'"{self.title}" has been successfully returned.'
        return f'"{self.title}" is already available.'

    def display(self):
        status = "Available" if self.available else "Borrowed"
        return f"Title: {self.title}, Author: {self.author}, Status: {status}"


class Library:
    def __init__(self):
        self.books = []

    def add_book(self, title, author):
        self.books.append(Book(title, author))

    def borrow_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                return book.borrow()
        return "Book not found."

    def return_book(self, title):
        for book in self.books:
            if book.title.lower() == title.lower():
                return book.return_book()
        return "Book not found."

    def display_books(self):
        if not self.books:
            print("No books in the library.")
        else:
            print("\nLibrary Books:")
            for book in self.books:
                print(book.display())


# Main Program
library = Library()

num_books = int(input("Enter the number of books to add: "))
for i in range(num_books):
    print(f"\nBook {i+1}")
    title = input("Enter title: ")
    author = input("Enter author: ")
    library.add_book(title, author)

library.display_books()

borrow_title = input("\nEnter the title of the book to borrow: ")
print(library.borrow_book(borrow_title))

library.display_books()

return_title = input("\nEnter the title of the book to return: ")
print(library.return_book(return_title))

library.display_books()
```

```python
# 2. Employee Payroll System

class Employee:
    def __init__(self, name, hourly_rate, hours_worked):
        self.name = name
        self.hourly_rate = hourly_rate
        self.hours_worked = hours_worked

    def calculate_salary(self):
        if self.hours_worked <= 160:
            base_salary = self.hourly_rate * self.hours_worked
            overtime_pay = 0
        else:
            base_salary = self.hourly_rate * 160
            overtime_hours = self.hours_worked - 160
            overtime_pay = overtime_hours * (self.hourly_rate * 1.5)

        total_salary = base_salary + overtime_pay
        return base_salary, overtime_pay, total_salary

    def display_payroll(self):
        base_salary, overtime_pay, total_salary = self.calculate_salary()
        print(f"\nEmployee: {self.name}")
        print(f"Hours Worked: {self.hours_worked}")
        print(f"Base Salary: {base_salary:.2f}")
        print(f"Overtime Pay: {overtime_pay:.2f}")
        print(f"Total Salary: {total_salary:.2f}")


# Main Program
employees = []

num_employees = int(input("Enter the number of employees: "))

if num_employees <= 0:
    print("At least one employee must be entered.")
else:
    for i in range(num_employees):
        print(f"\nEmployee {i+1}")
        name = input("Enter name: ")
        hourly_rate = float(input("Enter hourly wage: "))
        hours_worked = float(input("Enter total hours worked: "))

        if hourly_rate < 0 or hours_worked < 0:
            print("Invalid input: wage and hours cannot be negative.")
        else:
            employees.append(Employee(name, hourly_rate, hours_worked))

    print("\nPayroll Summary")
    for emp in employees:
        emp.display_payroll()
```

```python
# 3. Online Shopping Cart System

class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock


class ShoppingCart:
    def __init__(self):
        self.cart = {}

    def add_product(self, product, quantity):
        if quantity <= 0:
            return "Quantity must be greater than 0."
        if quantity > product.stock:
            return f"Cannot add {quantity} item(s). Only {product.stock} in stock."

        product.stock -= quantity
        if product.name in self.cart:
            self.cart[product.name]["quantity"] += quantity
        else:
            self.cart[product.name] = {
                "price": product.price,
                "quantity": quantity
            }
        return f"{quantity} {product.name}(s) added to cart."

    def remove_product(self, product, quantity):
        if product.name not in self.cart:
            return f"{product.name} is not in the cart."

        if quantity <= 0:
            return "Quantity must be greater than 0."

        if quantity >= self.cart[product.name]["quantity"]:
            removed_qty = self.cart[product.name]["quantity"]
            product.stock += removed_qty
            del self.cart[product.name]
            return f"All {product.name}(s) removed from cart."
        else:
            self.cart[product.name]["quantity"] -= quantity
            product.stock += quantity
            return f"{quantity} {product.name}(s) removed from cart."

    def view_cart(self):
        if not self.cart:
            print("\nCart is empty.")
            return

        total = 0
        print("\nCart Contents:")
        for name, info in self.cart.items():
            subtotal = info["price"] * info["quantity"]
            total += subtotal
            print(f"{name} - Price: {info['price']:.2f}, Quantity: {info['quantity']}, Subtotal: {subtotal:.2f}")
        print(f"Total Price: {total:.2f}")


# Predefined product list
products = [
    Product("Laptop", 50000, 5),
    Product("Mouse", 500, 10),
    Product("Keyboard", 1200, 8)
]

cart = ShoppingCart()

while True:
    print("\nAvailable Products:")
    for i, product in enumerate(products, start=1):
        print(f"{i}. {product.name} - Price: {product.price:.2f}, Stock: {product.stock}")

    print("\n1. Add Product")
    print("2. Remove Product")
    print("3. View Cart")
    print("4. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        product_num = int(input("Enter product number to add: "))
        if 1 <= product_num <= len(products):
            quantity = int(input("Enter quantity: "))
            print(cart.add_product(products[product_num - 1], quantity))
        else:
            print("Invalid product selection.")

    elif choice == "2":
        product_num = int(input("Enter product number to remove: "))
        if 1 <= product_num <= len(products):
            quantity = int(input("Enter quantity to remove: "))
            print(cart.remove_product(products[product_num - 1], quantity))
        else:
            print("Invalid product selection.")

    elif choice == "3":
        cart.view_cart()

    elif choice == "4":
        print("Exiting shopping cart system.")
        break

    else:
        print("Invalid choice.")
```
